future bike buyers to create your input for data mining algorithms. Remove all the unnecessary 
attributes. (Select features just by analysis). 

```python
df = pd.read_csv('./Kaggle/AwCustomers.csv')

col_remove = ['CustomerID', 'Title','FirstName', 'AddressLine1', 'LastName', 'MiddleName', 'Suffix' ,'AddressLine2', 'StateProvinceName', 'PostalCode', 'PhoneNumber', 'LastUpdated', 'CountryRegionName']
df = df.drop(columns=col_remove) 

df['BirthDate'] = pd.to_datetime(df['BirthDate']) 
today = pd.to_datetime('today')
df['Age'] = today.year - df['BirthDate'].dt.year
df['Age'] -= ((today.month < df['BirthDate'].dt.month) | ((today.month == df['BirthDate'].dt.month) & (today.day < df['BirthDate'].dt.day)))
df = df.drop(columns=['BirthDate'])

df.to_csv('selected.csv', index=False)
df.head()
```

Part II: Data Preprocessing and Transformation 
Use all the data rows (~= 18000 rows) with the selected features as input to apply all the tasks below, do

(a) Handling Null values

```python
df = pd.read_csv('selected.csv')
print(f"{df.isnull().sum()}")

null_cols = df.columns[df.isnull().sum() > 0]
print(f"Null Columns =  {null_cols}")
df['NumberCarsOwned'] = df['NumberCarsOwned'].fillna(df['NumberCarsOwned'].mean())
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])

print(f"{df.isnull().sum()}")
```

(b) Normalization

```python
numeric_cols = ['NumberCarsOwned', 'NumberChildrenAtHome', 'TotalChildren', 'YearlyIncome']
for i in numeric_cols:
    df[i] = (df[i] - df[i].min()) / (df[i].max() - df[i].min())
    print(f"Normalized {i} = \n{df[i]}")
```

(c) Discretization (Binning) on Continuous attributes or Categorical Attributes with too many different 
values

```python
# divide the range of YearlyIncome into 5 equal parts.
df['Income_Bin'] = pd.cut(df['YearlyIncome'], bins=4, labels=[1, 2, 3, 4])
print(f"Income_Bin = \n{df['Income_Bin']}")
```

(d) Standardization/Normalization

```python
for col in numeric_cols:
    df[col + '_std'] = (df[col] - df[col].mean()) / df[col].std()
    print(f"Std = \n{df[col + '_std']}")
```

(e) Binarization (One Hot Encoding)

```python
pd.set_option('display.max_columns', None)
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
arr = encoder.fit_transform(df[['Gender', 'MaritalStatus', 'Education', 'Occupation']]).toarray()
df = pd.DataFrame(arr, columns=encoder.get_feature_names_out(['Gender', 'MaritalStatus', 'Education', 'Occupation']))

df.head()
```

Part III: Calculating Proximity /Correlation Analysis of two features

(a) Calculate Similarity in Simple Matching, Jaccard Similarity, and Cosine Similarity between two 
following objects of your transformed input data

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
from numpy import dot
from numpy.linalg import norm

df = pd.read_csv('selected.csv')

num_cols = ['YearlyIncome','Age','NumberCarsOwned','NumberChildrenAtHome','TotalChildren','HomeOwnerFlag']
cat_cols = ['City','Education','Occupation','Gender','MaritalStatus']

scaler = StandardScaler()
numeric = scaler.fit_transform(df[num_cols])

encoder = OneHotEncoder(sparse_output=False)
nominal = encoder.fit_transform(df[cat_cols])

X = np.hstack([numeric, nominal])
obj1 = X[0]
obj2 = X[1]

simple_matching = np.sum(obj1==obj2)/len(obj1)
jaccard = np.sum((obj1>0)&(obj2>0))/np.sum((obj1>0)|(obj2>0))
cosine = dot(obj1,obj2)/(norm(obj1)*norm(obj2))

simple_matching, jaccard, cosine
print(f"Matching = {simple_matching}")
print(f"Jaccard = {jaccard}")
print(f"Cosine = {cosine}")
```
