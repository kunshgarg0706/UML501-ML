Q1. Write a Python program to scrape all available books from the website (https://books.toscrape.com/) Books to Scrape – a live site built for practicing scraping (safe, 
legal, no anti-bot)

```python
response = requests.get("https://books.toscrape.com/")
html = response.text
soup = BeautifulSoup(html, "html.parser")
```

1. Title 

```python
article = soup.find_all("article")

for i in article:
    print(f"Title = {i.h3.a["title"]}")
```

2. Price

```python
article = soup.find_all("p", class_="price_color")

for i in article:
    price = i.text
    print(f"Price = {price[1:len(price)]}")
```

3. Availability (In stock / Out of stock)

```python
article = soup.find_all("p", class_="availability")

for i in article:
    s = (i.text).strip()
    print(f"Availability = {s}")
```

4. Star Rating (One, Two, Three, Four, Five)

```python
article = soup.find_all("p", class_="star-rating")

for i in article:
    print(f"Star = {i["class"][1]}")
```

Combined

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = "https://books.toscrape.com/catalogue/page-{}.html"
titles = []
prices = []
availability = []
stars = []

page = 1
while True:
    url = base_url.format(page)
    response = requests.get(url)
    if response.status_code != 200:
        break
    soup = BeautifulSoup(response.text, "html.parser")
    articles = soup.find_all("article", class_="product_pod")
    if not articles:
        break
    for article in articles:
        titles.append(article.h3.a["title"])
        price = article.find("p", class_="price_color").text
        prices.append(price[1:len(price)])
        availability.append(article.find("p", class_="instock availability").text.strip())
        stars.append(article.find("p", class_="star-rating")["class"][1])
    page += 1

df = pd.DataFrame({"Title": titles, "Price": prices, "Availability": availability, "Star": stars})
df.to_csv("books.csv", index=False)
```

Q2. Write a Python program to scrape the IMDB Top 250 Movies list (https://www.imdb.com/chart/top/)

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

driver = webdriver.Chrome()
driver.get("https://www.imdb.com/chart/top/")

time.sleep(3)
movies = driver.find_elements(By.CLASS_NAME, "ipc-metadata-list-summary-item")
```

1. Rank (1–250)

```python
count = 1
rank = []
for i in movies:
    print(f"Rank = {count}")
    rank.append(count)
    count += 1
```

2. Movie Title 

```python
movie =[]
for i in movies:
    x = i.find_element(By.CSS_SELECTOR, "h3")
    print(f"Movie = {x.text}")
    movie.append(x.text)
```

3. Year of Release 

```python
year =[]
for i in movies:
    x = i.find_element(By.CLASS_NAME, "cli-title-metadata-item")
    print(f"Year of Release = {x.text}")
    year.append(x.text)
```

4. IMDB Rating

```python
rating = []
for i in movies:
    x = i.find_element(By.CSS_SELECTOR, ".ipc-rating-star--imdb").text.split("\n")[0]
    print(f"Rating = {x}")
    rating.append(x)
```

CSV

```python
df = pd.DataFrame({"Rank": rank, "Movie": movie, "Year of Release": year, "Rating": rating})
df.to_csv("movies.csv", index=False)
```

Q3. Write a Python program to scrape the weather information for top world cities from the 
given website (https://www.timeanddate.com/weather/)

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

response = requests.get("https://www.timeanddate.com/weather/")

html = response.text
soup = BeautifulSoup(html, "html.parser")
table = soup.find("table", class_="zebra")
rows = table.find_all("tr")[1:]
```

1. City Name

```python
city = []

for row in rows:
    city_td = row.find("a")
    if city_td:
        city.append(city_td.text.strip())
        print(f"City = {city_td.text.strip()}")
    else:
        city.append("N/A")
        print(f"City = N/A")
```

2. Temperature

```python
temp = []

for row in rows:
    temp_td = row.find("td", class_="rbi")
    if temp_td:
        temp.append(temp_td.text.strip())
        print(f"Temperature = {temp_td.text.strip()}")
    else:
        temp.append("N/A")
        print(f"Temperature = N/A")
```

3. Weather Condition (e.g., Clear, Cloudy, Rainy, etc.)

```python
weather = []

for row in rows:
    img_tag = row.find("img")
    if img_tag and "title" in img_tag.attrs:
        weather.append(img_tag["title"])
        print(f"Weather = {img_tag["title"]}")

    else:
        weather.append("N/A")
        print(f"Weather = N/A")
```

CSV

```python
import pandas as pd

print(len(city), len(temp), len(weather))

df = pd.DataFrame({"City": city, "Temperature": temp, "Weather": weather})
df.to_csv("weather.csv", index=False)
```
